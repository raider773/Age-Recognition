```python
valid_datagen = ImageDataGenerator(rescale=1./255)

valid_generator = valid_datagen.flow_from_dataframe(
        dataframe=data_val,
        directory=IMAGES_PATH,
        x_col='FileName',
        y_col='age',
        target_size=(200, 200),
        batch_size=256,
        class_mode='raw')
```

# Model

```python
from keras.layers import Dropout, Flatten, Dense, Activation, Lambda
import keras.backend as k
from keras.callbacks import ModelCheckpoint, LearningRateScheduler, EarlyStopping
from keras.models import Model 
from keras.applications import VGG16
from keras.constraints import max_norm
import keras
```

```python
model = VGG16(weights = "imagenet",
                           include_top = False,
                           input_shape = (200, 200, 3))

model.summary()
```

```python
for layer in model.layers:
    layer.trainable = False
```

```python
x = model.output
x = Flatten()(x)
x = Dense(200, activation="relu",kernel_constraint=max_norm(4))(x)
x = Dropout(0.2)(x)
x = Dense(200, activation="relu",kernel_constraint=max_norm(4))(x)
x = Dropout(0.2)(x)
```

```python
#branch expected value net

expected_branch = Dense(14,activation = 'softmax',name = 'softmax_age')(x)
lambda_layer = Lambda(lambda x: x * means, name = 'softmax_multiply')(expected_branch)
output_age= Lambda(lambda x: keras.backend.sum(x, axis=1,keepdims=True),name = 'output_age')(lambda_layer)
```

```python
model_final = Model(inputs = model.input, outputs = [output_age])
```

```python
model_final.summary()
```

```python
losses = {            
    'output_age': 'mse'
}

loss_weights = {      
    'output_age': 0.1    
}
```

```python
# callbacks 
checkpoint_file = "vgg16_1.h5"

checkpoint = ModelCheckpoint(checkpoint_file, monitor='val_loss', verbose=1, save_best_only=True,
                             save_weights_only=True, mode='auto', save_freq=1)

early = EarlyStopping(monitor='val_loss', min_delta=0, patience=5, verbose=1, mode='auto')
```

```python
import tensorflow as tf
def scheduler(epoch, lr):
    if epoch % 10 == 0:
        return lr * 0.1   
    return lr

callback_lr_decay = tf.keras.callbacks.LearningRateScheduler(scheduler)
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
batch_size = 256
learning_rate = 0.001
```

```python
model_final.compile(optimizer=Adam(learning_rate = learning_rate,beta_1 = 0.9,beta_2 = 0.9), loss=losses, metrics='mean_absolute_error', loss_weights=loss_weights)
```

```python
# Train the model 
hist = model_final.fit(train_generator,
                        steps_per_epoch = train_generator.n // batch_size,
                        epochs = 25,
                        validation_data = valid_generator,
                        validation_steps = valid_generator.n // batch_size,
                        callbacks = [callback_lr_decay],
                        verbose = 1)
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
```
